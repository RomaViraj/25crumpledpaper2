#### Activity 2: Data Cleaning

1. Make a new DataFrame with the following columns: 

  `['id', 'bairro', 'latitude', 'longitude', 'valor_prejuizo', 'Bicicleta', 'Bolsa ou Mochila', 'Carteira', 'Cartão de Crédito',  'Celular', 'Computador', 'DVD', 'Dinheiro', 'Documentos', 'Equipamento de Som', 'Estepe', 'MP4 ou Ipod', 'Móveis', 'Notebook', 'Outros', 'Relógio', 'Som', 'Tablet', 'Tv', 'time']`

2. Drop the `NaN` values from the `bairro` column and fill the `NaN` values of other columns with `False`.

3. Convert `True` and `False` into `0` and `1` by multiplying the values with 1. Calculate the total items stolen using the `sum()` function and drop the unnecessary columns.

Make a new DataFrame with the columns.

````python
# Create a variable with the column names in a list
features = ['id', 'bairro', 'latitude', 'longitude', 'valor_prejuizo',
       'Bicicleta', 'Bolsa ou Mochila', 'Carteira', 'Cartão de Crédito',
       'Celular', 'Computador', 'DVD', 'Dinheiro', 'Documentos',
       'Equipamento de Som', 'Estepe', 'MP4 ou Ipod', 'Móveis', 'Notebook',
       'Outros', 'Relógio', 'Som', 'Tablet', 'Tv', 'time']

# Create a new DataFrame with the columns selected
crime_df = df[features]
crime_df.head()
````

Drop the `NaN` values from `bairro` column using `dropna()` function. Fill `NaN` values of other columns with 0 using `fillna(False, inplace = True)`.


````python
# Drop the NaN from 'bairro' column.
crime_df.dropna(subset = ['bairro'], inplace = True)

# Fill the NaN values with False
crime_df.fillna(False, inplace = True)
````

Display the last 5 rows.

````python
# Display the last 5 rows of the DataFrame.
crime_df.tail()
````

Convert `True` and `False` into `0` and `1` by multiplying the values with 1 using `iloc[:,5:-1] * 1` function. 

Calculate the total items stolen using `iloc[:,5:-1].sum(axis = 1)` function and drop the unnecessary columns using `drop(columns = crime_df.iloc[:,5:-2].columns, inplace = True)`.

````python
# Convert True to 1 and False to 0
crime_df.iloc[:,5:-1] = crime_df.iloc[:,5:-1]*1

# Create a new column for total items stolen
crime_df['total_item_stolen'] = crime_df.iloc[:,5:-1].sum(axis = 1)

# Drop the unnecessary columns
crime_df.drop(columns = crime_df.iloc[:,5:-2].columns, inplace = True)
````

#### Activity 3: Visualizing the Crime Rates and Money Lost

1. Convert the `time` column into the datetime data type and extract the year.

2. Visualise the BRL (Brazilian Real Lost) and the total number of items stolen with different locations.

Convert the `time` column into datetime data type and extract the year storing it in a new column in a sorted manner

````python
# Convert the data type of 'time' column to datetime 
crime_df['time'] = pd.to_datetime(crime_df['time'])

# Extract year from 'time' column using 'year' attribute and 'DatetimeIndex' function of pandas. 
# Store the extracted year as a new column 'year' in the new DataFrame.
crime_df['year'] = pd.DatetimeIndex(crime_df['time']).year

# Sort the new DataFrame by year.
crime_df = crime_df.sort_values(by = "year")
````

Visualise the BRL (Brazilian Real) and the total number of items stolen with different locations by applying `density_mapbox()` function with `hover_data = ['valor_prejuizo', 'total_item_stolen']` to see both BRL and the total number of items stolen on the above `crime_df` DataFrame.

**Use following code to display the map**: 
```
fig = px.density_mapbox(crime_df, lat = 'latitude', lon = 'longitude', 
                        radius = 5, hover_data = ['valor_prejuizo', 'total_item_stolen'],
                        center = dict(lon = -46.6396, lat = -23.5558),
                        zoom = 8, mapbox_style='stamen-terrain',
                        color_continuous_scale = 'reds',

```

````python
# Import the required library
import plotly.express as px

# Create a figure object using parameters from the respective DataFrame
fig = px.density_mapbox(crime_df, lat = 'latitude', lon = 'longitude', 
                        radius = 5, hover_data = ['valor_prejuizo', 'total_item_stolen'],
                        center = dict(lon = -46.6396, lat = -23.5558),
                        zoom = 8, mapbox_style='stamen-terrain',
                        color_continuous_scale = 'reds',
                        animation_frame = 'year', width=1000, height=500,
                        title = 'BRL (Brazilian Real) Currency lost and Item Lost in São Paulo From 2010 to 2018')

fig.show()
````
